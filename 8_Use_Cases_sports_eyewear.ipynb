# Use Cases: Custom Fit Sports Eyewear


```python
import genysis
t = "eyJhbGciOiJIU.U892JiUy" # replace with your valid token
```

```python
# uploade required files to server
genysis.upload(src = "models/face_scan.obj", dest = "face_scan.obj", token = t)
genysis.upload(src = "models/support.obj", dest = "support.obj", token = t)
genysis.upload(src = "models/eyewear_frame.obj", dest = "eyewear_frame.obj", token = t)
```

```python
genysis.visualize("face_scan.obj",t,True,300,300) # a facial 3D scan
genysis.visualize("support.obj",t,True,300,300) # flat version of eyewear support structure 
genysis.visualize("eyewear_frame.obj",t,True,300,300) # fixed part of eyewear frame geometry
```

```python
# use cylindrical projection to create parametric surface at nose area of 3D facial scan
genysis.cylindricalProjection(
    target = "face_scan.obj",
    resolution = [40,20],
    output = "project_area.obj",
    token = t,
    center = [0,0,12], # center coordinate of projection cylinder
    height = 20, # height of projection cylinder
    range = [[0.225,0.275],[0,1]] # range of projection cylinder in u,v direction
)

# verify projection surface
genysis.visualize("project_area.obj",t,True)
```

```python
# apply same projection setting, this time map the flat support geometry to the parametric surface
genysis.cylindricalProjection(
    target = "face_scan.obj",
    resolution = [40,20],
    output = "support_projected.obj",
    token = t,
    center = [0,0,12], # center coordinate of projection cylinder
    height = 20, # height of projection cylinder
    range = [[0.225,0.275],[0,1]], # range of projection cylinder in u,v direction
    project_object = "support.obj",
    project_height = 2
)

genysis.visualize("support_projected.obj",t,True)
```

```python
# combine the projected support with fixed part of eyewear frame 
genysis.boolean(
    'eyewear_frame.obj',
    'support_projected.obj',
    'eyewear_final.obj',
    'union',
    t
)

genysis.visualize("eyewear_final.obj",t,True)
```
