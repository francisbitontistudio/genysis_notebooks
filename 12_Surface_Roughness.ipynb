# Surface Roughness



Surface Roughness function adds ramdomized beeds on surface of a given volume, which is useful in many medical implants applications

```python
import genysis
token = "Space0ut!"
```

```python
genysis.upload('models/shape.obj','shape.obj',token)
genysis.visualize('shape.obj',token,True)
```

```python
# create surface roughness class
s = genysis.surfaceRoughness()
# set geometry to add beeds on (has be to a closed volume), accepted formats are: obj,stl,stp,igs
s.setInput("shape.obj")
# set output file name, can be obj or stl
s.setOutput("surface_roughness.stl")

# OPTIONAL PARAMETERS:
# roughness size is the average distace of each beeds, by default will be 1/64 of geometry's bounding box
s.setRoughnessSize(2)
# set the radius of beeds, by default 1/2 of roughness size
s.setRadius(1.5)

# run the function
s.run(token)

genysis.visualize('surface_roughness.stl',token,True)
```
