## Example: Spine Surgical Guide


```python
import genysis
t = 'Space0ut!' # replace with your valid token
```

```python
# uploade required files to server
genysis.upload(src = "models/spine_part.obj", dest = "spine_part.obj", token = t)
genysis.upload(src = "models/surgical_guide.obj", dest = "surgical_guide.obj", token = t)
genysis.upload(src = "models/surgical_guide_hole.obj", dest = "surgical_guide_hole.obj", token = t)
```

```python
# visualize uploaded files
genysis.visualize("spine_part.obj",t,True,300,300) # part of spine MRI scan
genysis.visualize("surgical_guide.obj",t,True,300,300) # main geometry of surgical guide
genysis.visualize("surgical_guide_hole.obj",t,True,300,300) # volume to carve out of surgical guide for screw holes
```

```python
# use boolean difference to create a conformed surface between surgical guide and spine
genysis.boolean(
    input1 = "surgical_guide.obj",
    input2 = "spine_part.obj",
    output = "surgical_guide_midStep.obj",
    operation = "difference",
    token = t
)

genysis.visualize("surgical_guide_midStep.obj",t,True)
```

```python
# use boolean difference carve out the screw holes on surgical guide
genysis.boolean(
    input1 = "surgical_guide_midStep.obj",
    input2 = "surgical_guide_hole.obj",
    output = "surgical_guide_final.obj",
    operation = "difference",
    token = t
)

genysis.visualize("surgical_guide_final.obj",t,True)
```

```python
# downlod the finished result
genysis.download("surgical_guide_final.obj","models/surgical_guide_final.obj",t)
```
