# Use Cases: Generative Jewelry



```python
import genysis
t = "eyJhbGciOiJIU.U892JiUy" # replace with your valid token

# uploade required files to server
genysis.upload(src = "models/ring.obj", dest = "ring.obj", token = t)
genysis.upload(src = "models/ring_twisted.obj", dest = "ring_twisted.obj", token = t)
```

```python
genysis.visualize("ring.obj",t,True,300,300) # the original ring geometry
genysis.visualize("ring_twisted.obj",t,True,300,300) # a twisted ring geomtry
```

```python
# create blended version between original and twisted ring geometry
genysis.blend(
    "ring.obj", # input 1
    "ring_twisted.obj", # input2
    0.5, # blend value
    "ring_blended_0.5.obj", # output
    t # token
)

genysis.visualize("ring_blended_0.5.obj",t,True,300,300)
```

```python
# use boolean union to merge the 3 geometrys together
# for geometries which share overlaped faces, we strongly recommend to use voxel boolean engine
genysis.boolean(
    input1 = "ring.obj",
    input2 = "ring_blended_0.5.obj",
    operation = "union",
    output = "ring_midStep.obj",
    engine = "voxel", # use this to select voxel boolean engine
    vxsize = 0.05, # set voxel size, if skiped, voxel size will be set automatically as max(bounding-box-sides)/64
    token = t
)

genysis.boolean(
    input1 = "ring_midStep.obj",
    input2 = "ring_twisted.obj",
    operation = "union",
    output = "ring_final.obj",
    engine = "voxel",
    vxsize = 0.05,
    token = t
)

genysis.visualize("ring_final.obj",t,True)
```

```python
# download final result
genysis.download("ring_final.obj","models/ring_final.obj",t)
```
