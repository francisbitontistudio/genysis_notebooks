## Surface Lattice Custom Unit



```python
# install genysis library with: pip install genysis
import genysis
token = "eyJhbGciOiJIU.U892JiUy" # replace with your valid token

# upload files to be used in this demo
genysis.upload("models/surface_10by10.obj","surface_10by10.obj",token)
genysis.upload("models/chain_unit.obj","chain_unit.obj",token)
```

```python
# use custom-made obj file as unit for lattice obj file 
genysis.visualize("chain_unit.obj",token,True,300,300)
```

```python

surfaceLattice = genysis.surfaceLattice()
# by default the function auto scale the input unit to fix a 1x1x1 bounding box before arraying it to surface
# set auto scale to False here to allow unit geometry exceed this range to create a chainmail-style lattice
surfaceLattice.setAutoScale(False) 
surfaceLattice.setSurface("surface_10by10.obj")
surfaceLattice.setCellHeight(2)
surfaceLattice.setComponent("chain_unit.obj")
surfaceLattice.setOutput("surfaceLattice_chainmail.obj")
surfaceLattice.run(token)
genysis.visualize("surfaceLattice_chainmail.obj",token,True)
```
